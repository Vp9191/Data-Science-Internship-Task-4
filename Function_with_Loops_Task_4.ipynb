```python
#Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge amount must be at least ₹50"
    elif amount not in valid_plans:
        return "Invalid plan selected"
    else:
        return "Recharge Successful"


while True:
    amount = int(input("Enter recharge amount: "))
    result = validate_recharge(amount)
    print(result)

    if result == "Recharge Successful":
        break
    else:
        print("Please try again\n")
```

```python
#Inventory Reorder Alert System

def check_inventory(products):

    for product, quantity in products.items():

        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


products = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

check_inventory(products)
```

```python
#Student Result Processing System

def process_result(marks):

    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 55, 40, 70, 65]

result = process_result(student_marks)
print("Result:", result)
```

```python
#Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):

    fare = 50 + (distance * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:

    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", fare)

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() != "yes":
        break
```

```python
#Employee Attendance Eligibility Checker

def check_attendance(attendance):

    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]

result = check_attendance(attendance_list)
print("Status:", result)
```

```python
#Password Strength Checker

def check_password(password):

    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Password too short"

    for char in password:

        if char.isdigit():
            has_digit = True

        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
result = check_password(password)

print(result)
```
